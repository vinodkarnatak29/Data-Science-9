```python
#Read the dataset from the below link
#https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv

#Import Pandas Libraris
import pandas as pd

#Read the file with read_csv()
csvData=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')

#first 10 rows 
csvData.head(10)
```

```python
#1. Delete unnamed columns

result = csvData.drop(labels='Unnamed: 0',axis=1)

#first 10 rows
result.head(10)
```

```python
#2. Show the distribution of male and female

round(csvData['Gender'].value_counts(normalize=True)*100,2)
```

```python
#3. Show the top 5 most preferred names
# will use sort_values() for sorting and set ascending false to get in descending order.

csvData[["Name","Count"]].sort_values("Count",ascending=False).head(5)
```

```python
#4. What is the median name occurence in the dataset

#Name and Count column from Dataset
medianName = csvData[["Name","Count"]].groupby('Name').sum()

#Median name occurence in the dataset
medianName[medianName['Count'] == medianName['Count'].median()]
```

```python
#5. Distribution of male and female born count by states

#state gender distribution based on multi index
stateGenderDistribution= csvData[['State','Gender','Count']].set_index(['State', 'Gender'])

#Distribution of male and female born count by states
stateGenderDistribution.groupby(['State','Gender']).sum()
```
